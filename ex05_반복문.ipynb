#### 반복문
- 프로그램 내에서 똑같은 명령을 일정 횟수만큼 반복하여 수행하도록 제어하는 명령문
- 반복문 종류는 while문, for문이 있다.

```python
# while문
num1, num2 = 1,1

while(num1 != 0 or num2 != 0):
    num1 = int(input("첫 번재 정수 입력 >> "))
    num2 = int(input("두 번재 정수 입력 >> "))
    sum = num1 + num2
    if (num1 == 0 and num2 == 0):
        break;
    print("두 정수의 합 : %d" %(sum))
print("프로그램이 종료되었습니다.")
```

#### for문
 - 문자열 또는 리스트 또는 튜플이 들어갔을 때 안에 있는 요소를 하나씩 반복
 - for 변수 in 문자열(or 리스트 or 튜플):
     print(변수)

```python
list_food = ['삼겹살','김치볶음밥','파스타']
for i in list_food:
    print(i)
```

```python
str1 = '안녕하세요'
for s in str1:
    print(s)
```

```python
tuple_food = ('삼겹살','김치볶음밥','파스타', '짬뽕')
for food in tuple_food:
    print(food)
```

```python
score_list = [90, 45, 70, 60, 55]
for i in range(len(score_list)):
    if (score_list[i] >= 60):
        print("%d번 학생은 합격입니다." %(i+1))
    else:
        print("%d번 학생은 불합격입니다." %(i+1))
```

```python
score_list = [90, 45, 70, 60, 55, 90]
for i in score_list:
    if (i >= 60):
        print("{}번 학생은 합격입니다.".format(score_list.index(i)+1))
    else:
        print("{}번 학생은 불합격입니다.".format(score_list.index(i)+1))
```

#### for문
 - 필요한 만큼의 숫자를 만들어내는 유용한 기능
 - range(시작할 숫자, 종료할 숫자, 증가량)
 - range(1, 10, 1) -> 1부터 9까지 1씩 증가
 - range(1, 100, 3) -> 1부터 99까지 3씩 증가
 - range(10, 1, -1) -> 10부터 2까지 1씩 감소(-1씩 증가)

```python
for i in range(1, 11, 1):
    print(i,end='-')
```

```python
for i in range(1,5,1):
    print(i)
for j in range(1,5):
    print(j)
```

```python
for i in range(7):
    print(i)
```

```python
for i in range(0, 7, 2):
    print(i, end=' ')
```

```python
start = int(input("첫 번째 정수 입력 >> "))
end = int(input("두 번째 정수 입력 >> "))
for i in range(start, end+1):
    print(i, end=' ')
```

#### 리뷰 길이 평균 값


```python
sum = 0
for i in range(len(movie_review)):
    sum += len(movie_review[i][0])
average = sum / len(movie_review)
print("리뷰의 평균 길이는 '{}'입니다.".format(average))
```

```python
##### for-which 문
total = 0
for txt, pn in movie_review:
    total += len(txt)
print("리뷰의 평균 길이는 '{}'입니다.".format(total/len(movie_review)))
```

##### 긍정리뷰/부정리뷰 개수


```python
p = 0
n = 0
for txt, pn in movie_review:
    if (pn == 0):
        n += 1
    elif (pn == 1):
        p +=1
print("긍정 '{}'개\n부정 '{}'개'".format(p, n))
```

```python
##### 데이터 파악
pnTotal = 0
for txt, pn in movie_review:
    pnTotal += pn
print(f"긍정 '{pnTotal}'개")
print(f"부정 '{len(movie_review)-pnTotal}'개")
```
