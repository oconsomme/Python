##### 인덱싱(indexing)
- 무엇인가를 '가리킨다'는 의미

##### 슬라이싱(Slicing)
- 무엇인가를 '잘라낸다'는 의미

```python
s = "My name is MH"
```

```python
# 인덱싱
print(s[3])
print(s[-1])
```

```python
# 슬라이싱
# 변수명[start:end]
# 시작 인덱스가 0이거나 끝나는 인덱스가 -1일 경우 생략 가능
print(s[0:2])
print(s[:7])
print(s[8:])
print(s[:])
```

```python
# 문자열 안에 특정 값을 바꿔야 할 경우
month = 7
day = 3
# 문자열과 숫자형 데이터는 더하기 연산이 불가
# s = '오늘은 '+month+'월 '+day+'일 입니다.'
# print(s)
```

#### 문자열 포맷 코드
- %기호 사용


```python
s = '오늘은 %d월 %d일 입니다.'%(month, day)
print(s)
```

```python
# format 함수를 이용한 포매팅
s = '오늘은 {}월 {}일 입니다.'.format(month, day)
print(s)
```

```python
# f 문자열을 이용한 포매팅
s = f"오늘은 {month}월 {day}일 입니다."
print(s)
```

```python
x = 100
y = 200
sum2 = x+y
s1 = '%d와 %d의 합은 %d입니다.'%(x, y, sum2)
print(s1)

s2 = '{}와 {}의 합은 {}입니다.'.format(x,y,sum2)
print(s2)

s3 = f'{x}와 {y}의 합은 {sum2}입니다.'
print(s3)
```

#### 문자열 관련 함수


```python
s = "My name is MiHee"
```

```python
# 문자 개수 세기
# 문자열.count(데이터)
print(s.count('e'))
```

```python
# 위치 알려주기
# 문자열.find(데이터) : 찾는 문자가 존재하지 않을 경우 -1반환
# 문자열.index(데이터) : 찾는 문자가 존재하지 않을 경우 -에러발생
print(s.find('z'))
print(s.index('e'))
```

```python
# 문자열 삽입
# '문자'.join(문자열)
" ".join(s)
```

```python
# 문자열 나누기
# 문자열.split(구분문자)
s.split() # space tab enter 기준으로 나누기
```
