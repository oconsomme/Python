##### 실습문제


```python
score = int(input('점수 입력 >> '))
if 90 <= score <= 100 :
    print('{}점은 A학점 입니다.'.format(score))
elif score >= 80 :
    print('{}점은 B학점 입니다.'.format(score))
elif score >= 70 :
    print('{}점은 C학점 입니다.'.format(score))
elif score >= 60 :
    print('{}점은 D학점 입니다.'.format(score))
elif score < 60 :
    print('{}점은 F학점 입니다.'.format(score))
```

```python
print("-----[SM Coffee]-----")
print("금액을 입력하세요 >> ",end='')
money = int(input())
print("커피를 고르세요.\n[1]에스프레소(600원) [2]아메리카노(1000원) [3]카페라떼(1300원) >> ",end='')
select = int(input())
if select == 1 and money >= 600:
    money -= 600
elif select == 2 and money >= 1000:
    money -= 1000
elif select == 3 and money >= 1300:
    money -= 1300
else:
    print("돈이 부족해요..")
print("거스름돈: %d원" %(money))
```

```python
print('-----[SM Coffee]-----')
inputMoney = int(input('금액을 입력하세요 >> '))
print('커피를 고르세요.')
menu = int(input('[1]에스프레소(600원) [2]아메리카노(1000원) [3]카페라떼(1300원) >> '))

if menu == 1 :
    price = 600
elif menu == 2 :     
    price = 1000
elif menu == 3 : 
    price = 1300

if inputMoney < price :
    print('돈이 부족해요..T^T')
    price = 0
    
print(f'거스름돈 : {inputMoney - price}원')
```
